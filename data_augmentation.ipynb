```python
import os
import matplotlib.pyplot as plt

# Path to the directory
directory = "datasets/CK Original"
emotions = ["anger", "disgust", "fear", "happy", "sadness", "surprise", "contempt", "neutral"]

# Get the count of images in each subdirectory
emotions_counts = {emotion: 0 for emotion in emotions}
for subdir in os.listdir(directory):
    subdir_path = os.path.join(directory, subdir)
    if os.path.isdir(subdir_path):
        image_count = len(os.listdir(subdir_path))
        emotions_counts[emotions[int(subdir)]] = image_count

# Plot the bar chart
plt.bar(emotions_counts.keys(), emotions_counts.values())
plt.xlabel("Emotions")
plt.ylabel("Number of images")
plt.title("Number of images per emotion")
plt.xticks(rotation=45)
plt.show()
```

```python
from keras.preprocessing.image import ImageDataGenerator
import os

# Path to the directory
directory = "datasets/CK"
target_count = 300  # Desired number of images per folder

# Augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Augment images in each folder
for subdir in os.listdir(directory):
    subdir_path = os.path.join(directory, subdir)
    if os.path.isdir(subdir_path):
        image_count = len(os.listdir(subdir_path))
        if image_count < target_count:
            # Generate additional augmented images
            image_files = [os.path.join(subdir_path, file) for file in os.listdir(subdir_path)]
            image_generator = datagen.flow_from_directory(
                directory=directory,
                target_size=(48, 48),
                classes=[subdir],
                class_mode=None,
                batch_size=1,
                save_to_dir=subdir_path,
                save_prefix='augmented_'
            )
            for _ in range(target_count - image_count):
                image_generator.next()
```
